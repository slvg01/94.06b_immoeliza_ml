```python
import requests
from bs4 import BeautifulSoup

# Url of website
url = "https://www.immoweb.be/fr/recherche/maison/a-vendre/alost/9300?countries=BE&isALifeAnnuitySale=false&page=1&orderBy=relevance"
# I send my HTTP request with a "GET" to the site server to identify in the url
r = requests.get(url)
# I display the requested url and the return of the server
print(url, r.status_code)
# I will store the content of the website in a variable and print it
content = r.content
soup = BeautifulSoup(r.content, "html")
soup
```
